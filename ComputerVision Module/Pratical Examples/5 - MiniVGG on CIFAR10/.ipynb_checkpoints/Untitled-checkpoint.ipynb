# Mini VGG architecture on CIFAR10 dataset

The VGG architecture is a class of Convolutional Neural Networks. Proposed in 2014 by  [Simonyan and Zisserman](https://arxiv.org/pdf/1409.1556.pdf), their architecture reached the state of art, winning the ImageNet Challenge 2014. 

In this notebook, we introduce a smaller variation of VGG, with two blocks of convolution followed by the full connected layer to realize the classification. The main characteristic of VGG architecture is your kernel size during the convolution, a kernel of $(3x3)$ that we deploy here.

The dataset is CIFAR10 characterized by 10 classes with images of $(28x28)$.

## Importing Libraries

```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import classification_report
from compvis.ann.cnns import MiniVGG # compvis module
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

## Loading and prepocessing the dataset

```python
((X_train, y_train), (X_test, y_test)) = cifar10.load_data() # defining the training and test set
```

**Normalization**

```python
X_train = X_train.astype("float") / 255.
X_test = X_test.astype("float") / 255. # Normalization [0,1]
```

**Label Encoding**

```python
lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
y_test = lb.transform(y_test)
```

**List of targets**

```python
labelNames = ["airplane", "automobile", "bird", "cat", "deer","dog", "frog",
              "horse", "ship", "truck"]
```

## Building the model

We build the Mini VGG model with the compvis module, using the submodule cnns, we call MiniVGG class. Beyond the model, we also consider the SGD regularization.

**The Mini VGG** considered here is constituted by 4 convolutional layers, two with $32$ and $16$ filters.

```python
opt = SGD(lr = 0.1, decay = 0.01 / 40, momentum=0.9, nesterov=True) # regularization
model = MiniVGG.build(32, 32, 3, 10) # w, h, depth, classes
model.compile(loss = "categorical_crossentropy",
              optimizer = opt,
              metrics = ["accuracy"]) # Compiling the model
```

**Training the model with $40$ epochs**

```python
H = model.fit(X_train, y_train, validation_data = (X_test, y_test),
              batch_size = 64, epochs = 40, verbose = 1)
```

### Predicting on the trained model and evaluating it

**Predicting**

```python
predictions = model.predict(X_test, batch_size = 64)
```

**Classification Report**

```python
cr = classification_report(y_test.argmax(axis = 1), predictions.argmax(axis =1),
                           target_names = labelNames)
```

```python
print(cr)
```

**Visualizing the metrics results**

```python
plt.style.use("ggplot")
plt.figure(figsize=(20,20))
plt.plot(np.arange(0, 40), H.history["loss"], label="train_loss")
plt.plot(np.arange(0, 40), H.history["val_loss"], label="val_loss")
plt.plot(np.arange(0, 40), H.history["accuracy"], label="train_acc")
plt.plot(np.arange(0, 40), H.history["val_accuracy"], label="val_acc")
plt.title("Training Loss and Accuracy on CIFAR10")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend()
plt.show()
```

## Conclusion

Until here, our best score on the CIFAR10 dataset was $62\%$ using the Shallow Net. Using the Mini VGG we was capable to obtain an accuracy of $80\%$ on the test set and $85\%$ on the training set. This gap between the two curves can indicate over-fitting situation. But the important fact here, is to note how some regularization methods as Dropout, Batch Normalization and SGD, can improve considerably the model. And we can also see, how more deep is our CNN, it can bring some advantages in the final results. The results are far from the state of art reached by the authors, but are satisfactory.
